# Explanation

With the way things are, the truth set is too large to post to GitHub. I will therefore break it apart and put it back together again to get around this.

# Splitting the Truth Set

```python
import pandas as pd
```

### Test Case

```python
test_df = pd.DataFrame({"i":[i for i in range(20)], "j":[j for j in range(20, 40)]})
```

```python
test_df_1 = test_df[::4]
test_df_1
```

```python
test_df_2 = test_df[1::4]
test_df_2
```

```python
test_df_3 = test_df[2::4]
test_df_3
```

```python
test_df_4 = test_df[3::4]
test_df_4
```

```python
test_df_assembled = pd.concat([test_df_1, test_df_2, test_df_3, test_df_4])
test_df_assembled.sort_index()
```

As you can see, this method is perfectly fine to split the data up and they can easily be assembled again.

### Applying this to Our Truth Data

```python
whole_truth = pd.read_csv("truth_set.csv")
whole_truth
```

Splitting into 4 groups is still too big, but splitting into 5 is the least number of groups that I can create without exceeding the data limit.

```python
whole_truth_1 = whole_truth[::5]
whole_truth_2 = whole_truth[1::5]
whole_truth_3 = whole_truth[2::5]
whole_truth_4 = whole_truth[3::5]
whole_truth_5 = whole_truth[4::5]

split_list = [
    whole_truth_1, 
    whole_truth_2,
    whole_truth_3,
    whole_truth_4,
    whole_truth_5
]
```

```python
for i in range(5):
    whole_number = str(i+1)
    df_file_name = "data_chunks/split_truth_" + whole_number + ".csv"
    df_interest = split_list[i]
    df_interest.to_csv(df_file_name, index = False)
```

These data chunks are now okay in size to upload to GitHub.
